zee score example height/weight , applying zeescore in height column

outlier means to bring in certain range

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/sample_data/weight-height.csv")
df.head(5)
df.shape
```

```python
df.Height.describe()
```

zscore value should be between 1-3

```python
df['Zscore']= (df.Height - df.Height.mean()) / df.Height.std()
df.head(5)
```

finding outliers

```python
df[df['Zscore']>3]
```

```python
df[df['Zscore']< -3]
```

```python
no_outliers = df[(df.Zscore > -3) & (df.Zscore < 3)]
no_outliers.head(5)
```
